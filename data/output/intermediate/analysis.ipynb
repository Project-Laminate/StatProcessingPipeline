```python
# load the csv new_normative_df_output.csv

import pandas as pd

# load the csv file
df = pd.read_csv('new_normative_df_output.csv')

# print the distribution of the last 12 columns of the dataframe 
print(df.iloc[:, -18:].describe())
```

```python
# plot the distribution of the last 18 columns of the dataframe
df.iloc[:, -18:].hist(bins=50, figsize=(20, 15))
```

```python
# filter the data based on sex = F and age = +- 64 


# Calculating age range for filtering
age_range_min = 64*12 - 60  # Subtracting 5 years in months
age_range_max = 64*12 + 60  # Adding 5 years in months

# Filtering normative_df based on sex and age range
filtered_normative_df = df[
    (df["sex"] == 'F')
    & (df["age"] >= age_range_min)
    & (df["age"] <= age_range_max)
]
```

```python
# plot the distribution of the last 18 columns of the dataframe
filtered_normative_df.iloc[:, -18:].hist(bins=50, figsize=(20, 15))
```

```python
# plot the distribution of Left-Inf-Lat-Vent and Right-Inf-Lat-Vent 
filtered_normative_df[['Left-Inf-Lat-Vent', 'Right-Inf-Lat-Vent']].hist(bins=50, figsize=(10, 5))
```

```python
from scipy.stats import percentileofscore

percentile = percentileofscore(filtered_normative_df['Left-Inf-Lat-Vent'], 162.6, "rank")
print(f"Percentile of Left-Inf-Lat-Vent: {percentile}")
```
